## Problem

$$f(w_1, w_2) = 0.1w_1^2 + 2w_2^2$$

```python
import numpy as np
```

### GD

```python
def df_w(w):
    dW = np.array((0.2*w[0], 4*w[1]))
    return dW
```

```python
def sgd(W, dW, lr):
    W = W - lr*dW
    return W
```

```python
def train_p1(optimizer, lr, epochs):
    # initial point
    W = np.array([-5, -2], dtype=np.float32)
    results = [W]
    for e in range(epochs):
        dW = df_w(W)
        W = optimizer(W, dW, lr)
        results.append(W)
    return results
```

```python
train_p1(sgd, lr=0.4, epochs=30)
```
