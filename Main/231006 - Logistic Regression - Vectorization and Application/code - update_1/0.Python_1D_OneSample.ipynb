```python
import numpy as np
import matplotlib.pyplot as plt
import math
#%matplotlib qt

iris = np.genfromtxt('/content/drive/Othercomputers/My Laptop/Advanced/D-Learning/231006/code - update_1/iris_1D.csv',
                     dtype=None,
                     delimiter=',',
                     skip_header=1)
x_data = iris[:, 0]
y_data = iris[:, 1]

print(x_data)
print(y_data)
```

```python
def sigmoid_function(z):
    return 1 / (1 + math.exp(-z))

def loss_function(y_hat, y):
    return -y*math.log(y_hat) - (1 - y)*math.log(1 - y_hat)

def predict(x, w, b):
    z = w*x + b
    print('z: ', z)
    y_hat = sigmoid_function(z)

    return y_hat

def compute_gradient(x, y_hat, y):
    db = (y_hat - y)
    dw = x*(y_hat - y)

    return db, dw

def update(b, w, lr, db, dw):
    b = b - lr*db
    w = w - lr*dw

    return b, w
```

```python
# some parameters
lr = 0.01
num_epochs = 1

w = -0.1
b = 0.1

# for monitoring
losses = []

for _ in range(num_epochs):
    for i in range(4):
        # get data
        x = x_data[i]
        y = y_data[i]
        print('x, y:', x, y)

        # predict
        y_hat = predict(x, b, w)
        print('y_hat:', y_hat)

        # compute loss
        loss = loss_function(y_hat, y)
        print('loss:', loss)
        losses.append(loss)

        # compute gradient
        db, dw = compute_gradient(x, y_hat, y)
        print('db, dw:', db, dw)

        # update
        b, w = update(b, w, lr, db, dw)
        print('b, w:', b, w)
        #print('----------')
        break
```
